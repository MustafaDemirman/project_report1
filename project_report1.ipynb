```python
#First of all we have to write "conda install pandas" on Anaconda Prompt
#secondly we have to write "pip install nba_api" on Anaconda Prompt
#Team name : Duo of NBA

# Team members :

# Mustafa Demirman 201211018
# Burak Alım 201211301
# Title : NBA Data Analyst
 
 

# Data : From www.nba.com 

#Description : Getting team and player data from nba.com and  making analysis out of it about team and player performances.

#Goals : Detecting trendings about mainly teams and players from their win-lose ratio and win-lose strikes.

#Tools : Pandas,Matplotlib,requests,nba_api,msgpack,PyHamcrest
from nba_api.stats.static import teams

nba_teams = teams.get_teams()
# for example select the dictionary for the Celtics, which contains their team ID
celtics = [team for team in nba_teams if team['abbreviation'] == 'BOS'][0]
celtics_id = celtics['id']
```

```python
from nba_api.stats.endpoints import leaguegamefinder

# Query for games where the Celtics were playing
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=celtics_id)
# The first DataFrame of those returned is what we want.
games = gamefinder.get_data_frames()[0]
games.head()
```

```python
games.groupby(games.SEASON_ID.str[-4:])[['GAME_ID']].count().loc['2015':]
```

```python
# Subset the games to when the last 4 digits of SEASON_ID was 2017.
games_1718 = games[games.SEASON_ID.str[-4:] == '2017']
games_1718.head()
```
